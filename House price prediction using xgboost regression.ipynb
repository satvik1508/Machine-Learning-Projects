```python
#!pip install xgboost
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
df=pd.read_csv('boston.csv')
df.head()
```

```python
df.rename(columns={"MEDV":"PRICE"},inplace=True)
df.head()
```

```python
#check the number of rows and columns in df
df.shape
```

```python
#check the number of missing values
df.isnull().sum()
```

```python
#statistical measures
df.describe()
```

## Correlation between different features

### 1.Posititve correlation
### 2.Negative correlation

```python
correlation = df.corr()
```

```python
#ploting heatmap to understand correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

### seperating data and target

```python
X = df.drop(['PRICE'],axis=1)
y = df['PRICE']
```

### splitting data into train and test split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

#### model training using XG Booost regression method

```python
#loading the model
model=XGBRegressor()
```

```python
#training the model
model.fit(X_train,y_train)
```

### prediction on training data

```python
train_data_prediction=model.predict(X_train)
```

```python
train_data_prediction
```

```python
#R squared error
score1=metrics.r2_score(y_train,train_data_prediction)

#mean absolute error
score2=metrics.mean_absolute_error(y_train,train_data_prediction)
```

```python
plt.scatter(y_train,train_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices VS Predicted Prices")
plt.show()
```

```python
print('R square error:',score1)
print('mean absolute error:',score2)
```

### Prediction on test data

```python
test_data_prediction=model.predict(X_test)
```

```python
plt.scatter(y_test,test_data_prediction)
plt.xlabel("Actual Prices(test data)")
plt.ylabel("Predicted Prices(test data)")
plt.title("Actual Prices VS Predicted Prices")
plt.show()
```

```python
#R squared error
score3=metrics.r2_score(y_test,test_data_prediction)

#mean absolute error
score4=metrics.mean_absolute_error(y_test,test_data_prediction)
```

```python
print('R square error:',score3)
print('mean absolute error:',score4)
```
